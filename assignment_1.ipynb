```python
import requests
from bs4 import BeautifulSoup
```

```python
# 뉴스 데이터 수집 함수 정의
def get_news_data(keyword, total_news=50):
    news_list = []
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    for start in range(1, total_news + 1, 10):
        url = f"https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={keyword}&start={start}"
        res = requests.get(url, headers=headers)
        soup = BeautifulSoup(res.text, "html.parser")
        articles = soup.select('div[data-sds-comp="Profile"]')
        for article in articles:
            if len(news_list) >= total_news:
                break
            article_box = article.parent
            title_element = article_box.select_one('a[data-heatmap-target=".tit"]')
            title = title_element.get_text(strip=True) if title_element else "제목없음"
            press_element = article_box.select_one(
                "span.sds-comps-profile-info-title-text"
            )
            press = (
                press_element.get_text(strip=True) if press_element else "언론사없음"
            )
            summary_element = article_box.select_one(
                'span[class="sds-comps-text sds-comps-text-ellipsis sds-comps-text-ellipsis-3 sds-comps-text-type-body1"]'
            )
            summary = summary_element.get_text(strip=True) if summary_element else None
            news_list.append({"title": title, "press": press, "summary": summary})
    return news_list
```

```python
# 수집한 데이터 확인용 코드
def print_get_news_data(keyword):
    result = get_news_data(keyword)

    for idx, news in enumerate(result, 1):
        print(f"[{idx}] {news['title']} ({news['press']})")
        print(f" - {news['summary']}\n")
```

```python
print_get_news_data(keyword="코스피")
```

```python
print_get_news_data(keyword="코스닥")
```

```python
# 수집한 데이터를 pandas를 통해 데이터프레임으로 전환
import pandas as pd

kospi_list = get_news_data(keyword="코스피")
kosdaq_list = get_news_data(keyword="코스닥")

df_kospi = pd.DataFrame(kospi_list)
df_kospi["keyword"] = "코스피"

df_kosdaq = pd.DataFrame(kosdaq_list)
df_kosdaq["keyword"] = "코스닥"
```

```python
df_kospi.head(10)
```

```python
# 두 키워드 데이터프레임을 행방향 결합
df_total = pd.concat([df_kospi, df_kosdaq], ignore_index=True)

print(f"전체 수집 건수: {len(df_total)}")
display(df_total.head())
```

## 📊 결측치 탐지 및 처리 프로세스

---

### 1. 결측치 유형의 이론적 배경

- MCAR(완전 무작위 결측): 데이터의 누락이 다른 변수나 데이터 자체와 아무런 상관관계 없이 발생한 경우.
- MAR(무작위 결측): 결측 여부가 관측된 다른 변수(X)에 의해 설명될 수 있는 경우.
- MNAR(비무작위 결측): 결측 원인이 누락된 값(Y)자체에 있거나 데이터셋 외부 요인에 의한 경우.
    > 참고: 7강 '결측치 탐지'

---

### 2. 수집한 데이터의 분석 및 유형 판정

> **[판정 결과] MAR(Missing At Random)**

- 관측 데이터와의 연관성: 수집된 데이터프레임 내 `title(제목)`과 `press(언론사)` 변수는 모두 정상적으로 관측되었습니다.
- 추론: `summary(요약)`에 결측이 발생한 행들을 분석한 결과, 주로 메인 기사에 부속되어 제목에 특정 기호(... 등)가 포함된 '하위 관련 뉴스' 형태임을 확인했습니다.
- 판정 근거: 즉, 제목의 형태라는 기 관측된 정보(X)를 통해 요약문(Y)의 부재를 예측할 수 있으므로, 본 결측치는 MAR 메커니즘을 따르는 것으로 판단됩니다.

---

### 3. 데이터 정제 전략

본 과제의 목적은 키워드별 뉴스 요약문의 텍스트 분석입니다.

### 처리 방법: 완전 제거(Listwise Deletion)

| 분석 항목                 | 상세 내용                                                                                                                                                |
| :------------------------ | :------------------------------------------------------------------------------------------------------------------------------------------------------- |
| 결측치 비율에 대한 고찰   | 수집한 100개의 데이터 중 결측 비율이 57%(과제 작성 당시)로 매우 높으나, 본 데이터는 수치형이 아닌 비정형 텍스트(Text)라는 특수성이 있음.                 |
| 대치 기법 적용의 불가능성 | 요약문(`summary`)은 텍스트 데이터로, 평균/중앙값/KNN 대치가 불가능하며, 임의의 상수로 채울 경우 **단어 빈도 분석 결과에 심각한 편향**을 초래함.          |
| 도메인 지식을 활용        | 결측 행은 메인 뉴스를 짧게 반복하거나 제목만 노출되는 '하위 뉴스 리스트'로 판단됨. 이를 제거하는 것이 정보의 손실 보다 데이터 중복 정제 측면에서 유리함. |

> **[결론]**
> 결측 비율이 높음에도 불구하고, 분석 데이터의 품질과 텍스트 분석의 특수성을 고려하여 결측치가 포함된 행을 모두 제거하는 **완전 제거** 방법을 최종 채택하였습니다.
>
> _향후 과제 수행 시, 최종 분석 대상(30건)의 2~3배수를 초기 수집 목표로 설정하여 데이터 가용성을 확보할 필요가 있음을 시사함._


```python
# dropna를 통한 결측치 행 완전 제거
df_clean = df_total.dropna(subset=["summary"])

print(f"제거 전 데이터 수: {len(df_total)}")
print(f"제거 후 데이터 수: {len(df_clean)}")
```
