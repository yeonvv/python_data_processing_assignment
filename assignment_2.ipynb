```python
import os
from dotenv import load_dotenv

import requests
import json

load_dotenv()

api_key = os.getenv("API_KEY")

url = "http://apis.data.go.kr/B552584/ArpltnStatsSvc/getMsrstnAcctoRDyrg"
params = {
    "serviceKey": api_key,
    "returnType": "json",
    "numOfRows": 100,
    "pageNo": 1,
    "inqBginDt": "20260301",
    "inqEndDt": "20260315",
    "msrstnName": "진천동",
}

res = requests.get(url, params)
res.encoding = "utf-8"

result = json.loads(res.text)
print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
import pandas as pd

items = result["response"]["body"]["items"]
df = pd.DataFrame(items)

print(df)
```

```python
df.info()
```

```python
cols_to_numeric = [
    "pm10Value",
    "pm25Value",
    "so2Value",
    "coValue",
    "no2Value",
    "o3Value",
]
for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.info()
```

```python
df["msurDt"] = pd.to_datetime(df["msurDt"])
df["day_type"] = df["msurDt"].dt.dayofweek.apply(lambda x: "주말" if x >= 5 else "평일")

display(df.head())
```
