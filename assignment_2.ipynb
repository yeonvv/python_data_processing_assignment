```python
import os
from dotenv import load_dotenv

import requests
import json

# Extract

load_dotenv()

api_key = os.getenv("API_KEY")

url = "http://apis.data.go.kr/B552584/ArpltnStatsSvc/getMsrstnAcctoRDyrg"
params = {
    "serviceKey": api_key,
    "returnType": "json",
    "numOfRows": 100,
    "pageNo": 1,
    "inqBginDt": "20260301",
    "inqEndDt": "20260315",
    "msrstnName": "진천동",
}

res = requests.get(url, params)
res.encoding = "utf-8"

result = json.loads(res.text)
print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
# ~Extract

import pandas as pd

items = result["response"]["body"]["items"]
df = pd.DataFrame(items)

print(df)
```

```python
df.info()
```

```python
# Transform

cols_to_numeric = [
    "pm10Value",
    "pm25Value",
    "so2Value",
    "coValue",
    "no2Value",
    "o3Value",
]
for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.info()
```

```python
# 파생 변수 생성

df["msurDt"] = pd.to_datetime(df["msurDt"])
df["day_type"] = df["msurDt"].dt.dayofweek.apply(lambda x: "주말" if x >= 5 else "평일")

display(df.head())
```

```python
import matplotlib.pyplot as plt

# Load: 정제된 데이터를 csv 파일로 저장
df.to_csv("air_quality_jinchen.csv", index=False, encoding="utf-8-sig")

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

# Insight: 시간대별(일별) 데이터 변동 추이 시각화
plt.figure(figsize=(12, 5))
plt.plot(df["msurDt"], df["pm10Value"], marker="o", color="g", label="pm10(미세먼지)")
plt.plot(df["msurDt"], df["pm25Value"], marker="s", color="r", label="pm25(미세먼지)")

plt.title("대구 진천동 미세먼지 농도 변동 추이 (2026.03.01~03.15)")
plt.xlabel("날짜")
plt.ylabel("농도(㎍/㎥)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```
